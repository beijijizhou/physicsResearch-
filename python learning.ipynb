```python
def file(fname):
    df1= pd.read_csv(fname+".txt",delimiter=' ',skiprows=1)
    df1.drop(df1.columns[len(df1.columns)-1], axis=1, inplace=True)
    df1.info()
    df1.describe()
    Hist(fname,df1)
def Hist(fname,df1): 
    for name in df1.columns:
        s=np.sin(df1[name])
        hist,bins = np.histogram(s,bins = range(-1,1,20))       
        c=np.cos(df1[name])
        hist,bins = np.histogram(c,bins = range(-1,1,20))        
        fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
        axs[0].set_title(" sin "+name+" histogram") 
        axs[1].set_title(" cos "+name+" histogram") 
        axs[0].hist(s, 100)
        axs[1].hist(c, 100)   
```

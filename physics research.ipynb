```python
 import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def sinHist(fname,df1):   
    for name in df1.columns:
        r=np.sin(df1[name])
        hist,bins = np.histogram(r,bins = range(-1,1,20)) 
        plt.title(fname+" sin "+name+" histogram")
        plt.hist(r,100)
        plt.xlabel("X-axis(represents cos(x))")
        plt.ylabel("Y-axis(represents the number of frequency)")
        plt.show()
def cosHist(fname,df1):   
    for name in df1.columns:
        r=np.cos(df1[name])
        hist,bins = np.histogram(r,bins = [-1,-0.5,0,0.5,1]) 
        plt.title(fname+" cos "+name+" histogram")
        plt.hist(r,100)
        plt.xlabel("X-axis(represents cos(x))")
        plt.ylabel("Y-axis(represents the number of frequency)")
        plt.show()
def file(fname):
    df1= pd.read_csv(fname+".txt",delimiter=' ',skiprows=1)
    df1.drop(df1.columns[len(df1.columns)-1], axis=1, inplace=True)
    sinHist(fname,df1)
    cosHist(fname,df1)
```

```python
N_points = 10000
n_bins = 200

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(10000) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
plt.xlabel("(represents cos(x))")
plt.ylabel("(represents the number of frequency)")
```
